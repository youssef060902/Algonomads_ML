```python
# Drop the 'id' column and keep only the 30 predictive features + diagnosis
df_features = df.drop(['id'], axis=1)  # Keep diagnosis for optional coloring

# Select only the 30 numerical features (exclude diagnosis for correlation matrix)
features = df.drop(['id', 'diagnosis'], axis=1)


# Compute the correlation matrix (matrice de corrélation de Pearson)
correlation_matrix = features.corr()

# Display the full correlation matrix as a table (optional)
print("\n" + "="*60)
print("CORRELATION MATRIX (30x30) - Top 10 highest absolute correlations:")
print("="*60)
# Get absolute correlations, unstack, and sort
corr_unstacked = correlation_matrix.abs().unstack() # Transforme la matrice carrée en une longue liste (Série Pandas) de paires
corr_sorted = corr_unstacked.sort_values(kind="quicksort", ascending=False)
# Remove self-correlations (1.0) and duplicates
top_corr = corr_sorted[corr_sorted < 1.0].drop_duplicates().head(10)
print(top_corr)
```

On constate que certaines caractéristiques géométriques du noyau cellulaire (rayon, périmètre, aire – surtout dans leurs versions « worst » et « mean ») sont extrêmement corrélées (> 0.98). Cela est normal car ces trois mesures décrivent pratiquement la même chose : la taille du noyau. Cette forte multicolinéarité est à garder en tête lors de l’interprétation des modèles (notamment la régression logistique/Softmax), même si elle n’empêche pas d’obtenir une excellente précision (~97-99 %).

# DATA PREPARATION

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# 1. Suppression des colonnes inutiles/parasites
#    - 'id' → identifiant, non prédictif
#    - 'Unnamed: 32' → colonne vide créée par erreur lors de l'export CSV
columns_to_drop = ['id']

# Vérifie si 'Unnamed: 32' existe → si oui, on la supprime aussi
if 'Unnamed: 32' in df.columns:
    columns_to_drop.append('Unnamed: 32')
    print("Colonne parasite 'Unnamed: 32' détectée et supprimée !")

df = df.drop(columns=columns_to_drop)

print(f"Colonnes après nettoyage : {list(df.columns)}")
print(f"→ 31 colonnes restantes : diagnosis + 30 features (parfait !)")

# 2. Encodage de la cible : B → 0, M → 1
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
print(f"Encodage effectué → Bénin = 0, Malin = 1")

# 3. Séparation X (30 features) et y (target)
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
print(f"Shape de X : {X.shape} → doit être (569, 30)")

# 4. STANDARDISATION (z-score) → ÉTAPE OBLIGATOIRE de l'article
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Standardisation appliquée :")
print(f"   Moyenne ≈ {X_scaled.mean():.10f}")
print(f"   Écart-type = {X_scaled.std():.2f}")
```

```python
from imblearn.over_sampling import SMOTE


# 5. Split 80 % train / 20 % test (aléatoire + stratifié)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.20,       # ← 20 % test
    random_state=42,
    stratify=y            # garde la proportion B/M
)

print(f"Train : {X_train.shape[0]} échantillons → Bénins={sum(y_train==0)}, Malins={sum(y_train==1)}")
print(f"Test  : {X_test.shape[0]} échantillons → Bénins={sum(y_test==0)}, Malins={sum(y_test==1)}")

# =============================================================================
# 6. APPLICATION DE SMOTE (seulement sur le train !)
# =============================================================================
print("\nApplication de SMOTE sur le train set...")
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print(f"Après SMOTE → Train équilibré : Bénins={sum(y_train_res==0)}, Malins={sum(y_train_res==1)}")
```

# MODELING

```python
from sklearn.linear_model import LogisticRegression

# =============================================================================
# 1. BASELINE MODEL : Régression Logistique (meilleur baseline interprétable)
# =============================================================================
print("BASELINE MODEL : Régression Logistique (L2)")

baseline = LogisticRegression(random_state=42, max_iter=1000)
baseline.fit(X_train_res, y_train_res)

y_pred_baseline = baseline.predict(X_test)
acc_baseline = accuracy_score(y_test, y_pred_baseline)

print(f"Accuracy baseline (30 features) : {acc_baseline*100:.2f}%")

feature_names = X.columns
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

# =============================================================================
# 2. EXTRACTION DES FEATURES LES PLUS IMPORTANTES
# =============================================================================
# Méthode 1 : Coefficients de la régression logistique (valeur absolue)
coef_abs = np.abs(baseline.coef_[0])
importance_logreg = pd.Series(coef_abs, index=feature_names).sort_values(ascending=False)

# =============================================================================
# 3. SÉLECTION DES 10 FEATURES LES PLUS IMPORTANTES
# =============================================================================
top_10_features = importance_logreg.head(10).index.tolist()
print("TOP 10 FEATURES SÉLECTIONNÉES (coefficients Régression Logistique) :")
print(top_10_features)

# Affichage propre
plt.figure(figsize=(10, 6))
importance_logreg.head(10).sort_values().plot(kind='barh', color='steelblue')
plt.title("Top 10 Features les plus importantes\n(Méthode 1 : Coefficients Régression Logistique)", fontsize=14)
plt.xlabel("Valeur absolue du coefficient (|β|)")
plt.tight_layout()
plt.show()

# =============================================================================
# 4. PRÉPARATION DES DONNÉES RÉDUITES (10 features seulement)
# =============================================================================
# Indices des colonnes sélectionnées
selected_idx = [feature_names.get_loc(f) for f in top_10_features]

X_train_reduced = X_train[:, selected_idx]
X_test_reduced  = X_test[:, selected_idx]

# Si tu as déjà appliqué SMOTE sur X_train → on applique aussi sur les données réduites
if 'X_train_res' in locals():
    X_train_reduced = X_train_res[:, selected_idx]
    y_train_to_use = y_train_res
    print(f"SMOTE déjà appliqué → données réduites équilibrées")
else:
    y_train_to_use = y_train
    print("SMOTE non appliqué → on garde le train original")

# =============================================================================
# 5. MODÈLE RÉDUIT : SOFTMAX REGRESSION (HYPERPARAMÈTRES EXACTS DE L'ARTICLE)
# =============================================================================
print("\n" + "="*80)
print("MODÈLE RÉDUIT : SOFTMAX REGRESSION (comme dans Agarap 2018)")
print("="*80)

tf.keras.backend.clear_session()

model_reduced = Sequential([
    Dense(2, input_shape=(10,), activation='softmax')  # 10 features → sortie softmax
])

model_reduced.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),   # ← EXACT 1e-3
    loss='sparse_categorical_crossentropy',                  # ← Cross-entropy
    metrics=['accuracy']
)

print(model_reduced.summary())

# Entraînement avec les hyperparamètres EXACTS de l'article
history = model_reduced.fit(
    X_train_reduced, y_train_to_use,
    batch_size=128,      # ← EXACT 128
    epochs=3000,         # ← EXACT 3000
    verbose=1
)
```

```python
# On réutilise X_train_res / y_train_res (déjà avec SMOTE) et X_test (30 features)
# → Pas besoin de refaire le split ni SMOTE

tf.keras.backend.clear_session()

# Modèle Softmax Regression sur les 30 features (EXACTEMENT comme dans l'article)
model_full = Sequential([
    Dense(2, input_shape=(30,), activation='softmax')  # 30 features → 2 classes
])

model_full.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),  # ← 1e-3 EXACT
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print(model_full.summary())

# Entraînement avec hyperparamètres IDENTIQUES à l'article
print("Entraînement en cours (3000 epochs, batch=128)...")
history_full = model_full.fit(
    X_train_res, y_train_res,      # ← données avec SMOTE
    batch_size=128,
    epochs=3000,
    verbose=0
)
```
