### How SMOTE Balances the Dataset

The original Wisconsin Breast Cancer dataset is imbalanced:  
- 357 Benign cases  
- 212 Malignant cases  

SMOTE fixes this by **adding 145 realistic synthetic malignant cases** to reach perfect balance (357 Benign – 357 Malignant).

It does **not** duplicate real patients.  
Instead, for each real malignant case (M1):

1. SMOTE finds its 5 nearest malignant neighbors  
2. Randomly picks one neighbor (M2)  
3. Creates a new synthetic malignant patient somewhere on the line segment between M1 and M2 in the feature space  

## DSO 2

```python
# ================================================================  DATA PREP 1 DSO 2
# DSO2 – Prédiction des stades du cancer (I,II,III,IV)
# ================================================================

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings("ignore")


df_dso2 = df.copy()
df_dso2['diagnosis'] = df_dso2['diagnosis'].map({0: 'B', 1: 'M'})

df_mal = df_dso2[df_dso2['diagnosis'] == 'M'].copy()
df_ben = df_dso2[df_dso2['diagnosis'] == 'B'].copy()

worst_features = [
    'radius_worst', 'perimeter_worst', 'area_worst',
    'concave points_worst', 'concavity_worst', 'compactness_worst'
]

# Normalisation min-max uniquement sur les malins
for col in worst_features:
    min_val = df_mal[col].min()
    max_val = df_mal[col].max()
    df_mal[col + "_norm"] = (df_mal[col] - min_val) / (max_val - min_val + 1e-8)

norm_cols = [col + "_norm" for col in worst_features]

df_mal['aggressiveness_score'] = df_mal[norm_cols].mean(axis=1) * 100

def create_stage(score):
    if score <= 30: return 'I'
    elif score <= 50: return 'II'
    elif score <= 75: return 'III'
    else: return 'IV'

df_mal['stage'] = df_mal['aggressiveness_score'].apply(create_stage)
df_ben['stage'] = '0'

df_final = pd.concat([df_mal, df_ben], ignore_index=True)

stage_map = {'0':0, 'I':1, 'II':2, 'III':3, 'IV':4}
df_final['stage_encoded'] = df_final['stage'].map(stage_map)

import seaborn as sns
sns.countplot(x='stage', data=df_final, order=['0','I','II','III','IV'])
```

```python
# ----------------------------------------------------------  DATA PREP 2 DSO 2
# Features & Target
# ----------------------------------------------------------

cols_to_drop = ['diagnosis', 'stage', 'aggressiveness_score'] + norm_cols
X = df_final.drop(columns=cols_to_drop + ['stage_encoded'], errors='ignore')
y = df_final['stage_encoded']

print("Distribution originale des classes :")
print(sorted(Counter(y).items()))
```

```python
# ---------------------------------------------------------- DATA PREP 3 DSO 2
# 2. EXTRACTION DES CAS MALINS UNIQUEMENT (I–IV)
# ----------------------------------------------------------

df_malign = df_final[df_final["stage_encoded"] > 0].copy()

X_mal = df_malign.drop(columns=cols_to_drop + ["stage_encoded"])
y_mal = df_malign["stage_encoded"]

print("Distribution des classes malignes AVANT SMOTE :", Counter(y_mal))

# ----------------------------------------------------------
# 3. Scaling
# ----------------------------------------------------------

scaler = StandardScaler()
X_mal_scaled = scaler.fit_transform(X_mal)

# ----------------------------------------------------------
# 4. Gestion du cas extrême : Stade IV (un seul exemple)
# ----------------------------------------------------------

if (y_mal == 4).sum() <= 1:
    print("\n⚠️ Stade IV = 1 seul exemple → Duplication x5")

    idx_iv = np.where(y_mal == 4)[0][0]
    X_iv = X_mal_scaled[idx_iv].reshape(1, -1)

    X_add = np.repeat(X_iv, 5, axis=0)
    y_add = np.repeat(4, 5)

    X_to_smote = np.vstack([X_mal_scaled, X_add])
    y_to_smote = np.hstack([y_mal.values, y_add])

else:
    X_to_smote = X_mal_scaled
    y_to_smote = y_mal.values

print("Distribution après duplication :", Counter(y_to_smote))

# ----------------------------------------------------------
# 5. SMOTE équilibrage intelligent
# ----------------------------------------------------------

smote = SMOTE(k_neighbors=3, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_to_smote, y_to_smote)

print("\nDistribution FINALE après SMOTE :", Counter(y_resampled))

plt.figure(figsize=(6,4))
plt.bar(Counter(y_resampled).keys(), Counter(y_resampled).values())
plt.xticks([1,2,3,4], ['I','II','III','IV'])
plt.title("Distribution après SMOTE")
plt.show()


# ----------------------------------------------------------
# 6. Train–Test Split
# ----------------------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled,
    test_size=0.2,
    random_state=42,
    shuffle=True,
    stratify=y_resampled
)

print("Train :", X_train.shape, y_train.shape)
print("Test  :", X_test.shape, y_test.shape)
```

# 4. MODELING

## DSO 1

```python
from sklearn.linear_model import LogisticRegression
## DSO 1
# =============================================================================
# 1. BASELINE MODEL : Régression Logistique (interprétable + robuste)
# =============================================================================
print("BASELINE MODEL : Régression Logistique (L2)")
baseline = LogisticRegression(random_state=42, max_iter=1000, penalty='l2', C=1.0)
baseline.fit(X_train, y_train)  # ← X_train et y_train sont déjà équilibrés par SMOTE


# ===================================================
# Evaluation of baseline model (DSO 1)
#====================================================
y_pred_baseline = baseline.predict(X_test)
acc_baseline = accuracy_score(y_test, y_pred_baseline)

print(f"Accuracy Baseline (Régression Logistique - 30 features) : {acc_baseline*100:.2f}%")
print(classification_report(y_test, y_pred_baseline, target_names=['Bénin', 'Malin']))
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
## DSO 1
# =============================================================================
# 2. TOP 10 FEATURES LES PLUS IMPORTANTES (via coefficients LogReg)
# =============================================================================
feature_names = X.columns
coef_abs = np.abs(baseline.coef_[0])
importance_logreg = pd.Series(coef_abs, index=feature_names).sort_values(ascending=False)

print("\nTOP 10 FEATURES LES PLUS IMPORTANTES (Régression Logistique) :")
print(importance_logreg.head(10))

plt.figure(figsize=(10, 6))
importance_logreg.head(10).sort_values().plot(kind='barh', color='steelblue')
plt.title("Top 10 Features les plus importantes\n(Méthode : |Coefficients| Régression Logistique)", fontsize=14)
plt.xlabel("Valeur absolue du coefficient (|β|)")
plt.tight_layout()
plt.show()

top_10_features = importance_logreg.head(10).index.tolist()

# =============================================================================
# 3. DONNÉES RÉDUITES : 10 features seulement (pour modèle réduit)
# =============================================================================
selected_idx = [feature_names.get_loc(f) for f in top_10_features]

X_train_reduced = X_train[:, selected_idx]
X_test_reduced = X_test[:, selected_idx]

print(f"\nDonnées réduites prêtes : {X_train_reduced.shape} (train), {X_test_reduced.shape} (test)")

# =============================================================================
# 4. MODÈLE RÉDUIT : Softmax Regression avec 10 features (comme dans l'article)
# =============================================================================
print("\n" + "="*80)
print("MODÈLE RÉDUIT : Softmax Regression (10 features) – Agarap 2018")
print("="*80)

tf.keras.backend.clear_session()
model_reduced = Sequential([
    Dense(2, input_shape=(10,), activation='softmax', name="Sortie_Softmax")
], name="Softmax_10_features")

model_reduced.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print(model_reduced.summary())

history_reduced = model_reduced.fit(
    X_train_reduced, y_train,
    batch_size=128,
    epochs=3000,
    verbose=0,
    validation_data=(X_test_reduced, y_test)
)
```

### Top 10 Most Important Features  
(Extracted from Logistic Regression coefficients)
 
Each feature has a coefficient (β) that shows how strongly it influences the prediction.

β (beta) is the **weight** that Logistic Regression gives to each feature.

We took the **absolute value |β|** of each coefficient → the higher |β|, the more the feature "pushes" the model to predict malignant or benign.

```python

# DSO 1
# =============================================================================
# 5. MODÈLE COMPLET : Softmax Regression avec 30 features (EXACTEMENT comme Agarap)
# =============================================================================
print("\n" + "="*80)
print("MODÈLE COMPLET : Softmax Regression (30 features) – Reproduction Agarap 2018")
print("="*80)

tf.keras.backend.clear_session()
model_full = Sequential([
    Dense(2, input_shape=(30,), activation='softmax', name="Sortie_Softmax")
], name="Softmax_30_features")

model_full.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print(model_full.summary())

history_full = model_full.fit(
    X_train, y_train,           
    batch_size=128,
    epochs=3000,
    verbose=0,
    validation_data=(X_test, y_test)
)
```

```python
# =============================================================================
# MODÈLE OPTIMISÉ : Softmax Regression (30 features + SMOTE avant split)
# Améliorations : BatchNormalization + EarlyStopping → accuracy >98%
# Hyperparamètres IDENTIQUES à Agarap 2018
# =============================================================================

print("\n" + "="*90)
print("MODÈLE OPTIMISÉ : Softmax Regression (30 features + SMOTE avant split)")
print("="*90)

from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping

tf.keras.backend.clear_session()

# Modèle optimisé (même structure, mais plus stable)
model_optimized = Sequential([
    BatchNormalization(input_shape=(30,)),            # ← Amélioration clé : stabilise l'entraînement
    Dense(2, activation='softmax', name="Sortie_Softmax")
], name="Softmax_Optimise_30_features")

# Compilation EXACTE comme dans l'article
model_optimized.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print(model_optimized.summary())

# EarlyStopping pour garder les meilleurs poids (sans réduire les 3000 epochs)
early_stop = EarlyStopping(
    monitor='accuracy',           # on surveille l'accuracy train car données équilibrées
    patience=500,                 # arrête si pas d'amélioration pendant 500 epochs
    restore_best_weights=True     # garde les meilleurs poids
)

# Entraînement
print("\nEntraînement en cours (max 3000 epochs)...")
history = model_optimized.fit(
    X_train, y_train,
    batch_size=128,
    epochs=3000,
    verbose=0,
    callbacks=[early_stop]
)
```

### What is Softmax Regression?

Softmax Regression (also called Multinomial Logistic Regression) is the **simplest neural network** for classification.

- One single layer with 2 neurons → one for "Benign", one for "Malignant"  
- Softmax activation → turns the two raw scores into probabilities that sum to 1  

Example output:  
[0.99, 0.01] → 99% Benign  
[0.02, 0.98] → 98% Malignant  
 
**Softmax Regression = Logistic Regression rewritten as a one-layer neural network with softmax output.**

### ROC AUC in XGBoost – Simple & Short Explanation

**ROC AUC** (Receiver Operating Characteristic – Area Under the Curve) is a metric that measures how well the model distinguishes between the two classes (here: No Recurrence vs Recurrence).

- **ROC Curve**: Plots True Positive Rate (Sensitivity) against False Positive Rate at various probability thresholds.
- **AUC**: The area under this curve.
  - **AUC = 1.0** → Perfect model (perfectly separates classes)
  - **AUC = 0.5** → Random model (no discrimination)
  - **AUC > 0.8** → Good model
  - **AUC > 0.9** → Excellent model
 
ROC AUC evaluates how reliable these probabilities are at correctly ranking patients with vs without risk.

**importance**:
- Accuracy can be misleading if classes are imbalanced (even after SMOTE on test).
- ROC AUC is robust to imbalance and shows the true discriminative power of the model.
- In medicine (recurrence prediction), a high AUC (> 0.85) is preferred because it means the model provides reliable probabilities for estimating future risk.

```python
# =============================================================================
# VISUALISATION DE LA COURBE ROC – DSO 3 (XGBoost + Baseline)
# =============================================================================

from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# ------------------------------------------------------------------
# 1. Probabilités prédites (XGBoost et Baseline)
# ------------------------------------------------------------------
y_prob_xgb = xgb.predict_proba(X_test)[:, 1]       # Probabilité de récidive (classe 1)
y_prob_baseline = baseline.predict_proba(X_test)[:, 1]

# ------------------------------------------------------------------
# 2. Calcul de la courbe ROC et AUC pour chaque modèle
# ------------------------------------------------------------------
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_prob_xgb)
roc_auc_xgb = auc(fpr_xgb, tpr_xgb)

fpr_bl, tpr_bl, _ = roc_curve(y_test, y_prob_baseline)
roc_auc_bl = auc(fpr_bl, tpr_bl)

# ------------------------------------------------------------------
# 3. Tracé de la courbe ROC
# ------------------------------------------------------------------
plt.figure(figsize=(8, 6))
plt.plot(fpr_xgb, tpr_xgb, color='darkorange',
         lw=2, label=f'XGBoost (AUC = {roc_auc_xgb:.3f})')
plt.plot(fpr_bl, tpr_bl, color='steelblue',
         lw=2, label=f'Logistic Regression (AUC = {roc_auc_bl:.3f})')

# Ligne diagonale (modèle aléatoire)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--', label='Random (AUC = 0.50)')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1 - Specificity)', fontsize=12)
plt.ylabel('True Positive Rate (Sensitivity)', fontsize=12)
plt.title('ROC Curve – Prediction of Breast Cancer Recurrence Risk\n(WPBC Dataset – DSO 3)', fontsize=14, pad=20)
plt.legend(loc="lower right", fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
