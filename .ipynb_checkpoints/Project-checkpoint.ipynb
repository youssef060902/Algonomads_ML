```python
# ============================================================
#                 DATA PREPARATION (WDBC)
# ============================================================

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

RND = 42

# ============================================================
# 2) Supprimer les colonnes inutiles
#    - id (pas utile)
#    - Unnamed: 32 (colonne vide dans ce dataset)
# ============================================================
# supprimer colonnes inutiles
df = df.drop(columns=["id", "Unnamed: 32"], errors='ignore')
# encoder target
df["diagnosis"] = df["diagnosis"].map({"B": 0, "M": 1})

# Features / target
X = df.drop("diagnosis", axis=1)
y = df["diagnosis"]

# train/test split 70/30 stratifié (comme l'article)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=RND, stratify=y
)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
# -------------------------
# 2) Pipeline + GridSearch
# -------------------------
# Pipeline: scaler -> (optionnel selectkbest) -> knn
pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("select", SelectKBest(score_func=f_classif)),  # on va gridsearch size k
    ("knn", KNeighborsClassifier())
])

# Grille à tester
param_grid = {
    # SelectKBest: None possible via k = "all" simulation: on met k = X.shape[1] pour tout garder
    "select__k": [5, 10, 15, 20, 25, X.shape[1]],

    # KNN params
    "knn__n_neighbors": [1, 3, 5, 7, 9],       # inclut 1 comme l'article
    "knn__weights": ["uniform", "distance"],   # distance weighting testé
    "knn__metric": ["euclidean", "manhattan"]  # L2 et L1
}

# Stratified CV
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RND)

# GridSearch (scoring sur accuracy)
grid = GridSearchCV(
    pipe,
    param_grid,
    cv=cv,
    scoring="accuracy",
    n_jobs=-1,
    verbose=1,
    return_train_score=True
)

# Lancer recherche
grid.fit(X_train, y_train)

# -------------------------
# 3) Résultats recherche
# -------------------------
print("\n-> Meilleurs paramètres trouvés (CV):")
print(grid.best_params_)
print("Best CV Accuracy: {:.4f}".format(grid.best_score_))

# -------------------------
# 4) Évaluation sur le test set
# -------------------------
best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("\nTest Accuracy (best model): {:.4f} -> {:.2f}%".format(acc, acc*100))

print("\nConfusion Matrix :")
cm = confusion_matrix(y_test, y_pred)
print(cm)

print("\nClassification Report :")
print(classification_report(y_test, y_pred, digits=4))



# -------------------------
# 5) Si accuracy < 0.95 -> propose et évalue alternative simple (1-NN euclidien)
# -------------------------
if acc < 0.95:
    print("\n⚠️ Accuracy < 95%. J'essaie maintenant 1-NN (euclidean) sans sélection de features ...")
    # pipeline simple: scaler -> knn(1, euclidean)
    from sklearn.pipeline import make_pipeline
    model_1nn = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=1, metric="euclidean", weights="uniform"))
    model_1nn.fit(X_train, y_train)
    y_pred_1nn = model_1nn.predict(X_test)
    acc_1nn = accuracy_score(y_test, y_pred_1nn)
    print("1-NN euclidean Test Accuracy: {:.4f} -> {:.2f}%".format(acc_1nn, acc_1nn*100))
    print("Confusion Matrix (1-NN):")
    print(confusion_matrix(y_test, y_pred_1nn))
    print("\nClassification Report (1-NN):")
    print(classification_report(y_test, y_pred_1nn, digits=4))
    if acc_1nn > acc:
        print("\n✅ 1-NN euclidean est meilleur que le modèle GridSearch. Nous allons l'utiliser.")
        best_model = model_1nn
        acc = acc_1nn

# -------------------------
# 6) Plot confusion matrix 
# -------------------------
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title(f"Confusion Matrix (Accuracy={acc*100:.2f}%)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()



# -------------------------
# 7) Résumé final
# -------------------------
print("\n=== Résumé final ===")
print("Test Accuracy finale : {:.4f} ({:.2f}%)".format(acc, acc*100))
print("Meilleurs paramètres finaux :")
try:
    # si GridSearch best_params_ exists
    print(grid.best_params_)
except:
    print("Best params not available (using fallback 1-NN).")
```

```python
df_staged= df.copy()

aggressive_features= ['radius_worst', 'concavity_worst','concave points_worst', 'area_worst', 'compactness_worst', 'perimeter_worst']

for feature in aggressive_features:
    min_val=df_staged[feature].min()
    max_val = df_staged[feature].max()
    df_staged[feature]=(df_staged[feature]-min_val)/(max_val-min_val)

cols=[feature for feature in aggressive_features]
df_staged['aggressiveness_score']=df_staged[cols].mean(axis=1)*100


def assign_morphology_stage(row):

    score = row['aggressiveness_score']
    radius = row['radius_worst']
    concavity = row['concavity_worst']
    diagnosis = row['diagnosis']
    
    if (score < 30) and (radius < 15) and (concavity < 0.1):
        return '1'
    
    elif (score > 80) or (diagnosis == 1 and (radius > 20 or concavity > 0.3)):
        return '4'
    
    elif (score > 60) and (diagnosis == 1):
        return '3'
    
    else:
        return '2'
    # Apply stage assignment
df_staged['morphology_stage'] = df_staged.apply(assign_morphology_stage, axis=1)

print("Morphological stages created successfully!")
print(f"\nStage Distribution:")
stage_counts = df_staged['morphology_stage'].value_counts().sort_index()
for stage, count in stage_counts.items():
    percentage = (count / len(df_staged)) * 100
    print(f"  Stage {stage}: {count} samples ({percentage:.1f}%)")


# Remove temporary normalized columns
# df_staged = df_staged.drop(norm_cols, axis=1)
```

```python
# Select features for the model 
from sklearn.calibration import LabelEncoder


X2 = df_staged.drop([ 'diagnosis', 'morphology_stage', 'aggressiveness_score'], 
                   axis=1, errors='ignore')
y2 = df_staged['morphology_stage']


# Convert stages to numbers 
le = LabelEncoder()
y_encoded = le.fit_transform(y2)
```

```python
# Split data into training (80%) and testing (20%)
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    X2, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)


# Scale features (important for most ML algorithms)
scaler = StandardScaler()
X_train_scaled2 = scaler.fit_transform(X_train2)
X_test_scaled2= scaler.transform(X_test2)

print("\nFeatures scaled (mean=0, variance=1)")
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# ------------------------
# 1) Pipeline complet
# ------------------------

pipe = Pipeline([
    ("select", SelectKBest(score_func=f_classif)),
    ("knn", KNeighborsClassifier())
])

param_grid = {
    "select__k": [10, 15, 20, 25, 30],   # top features
    "knn__n_neighbors": [1, 3, 5, 7, 9, 11, 15, 21, 31],
    "knn__weights": ["uniform", "distance"],
    "knn__metric": ["euclidean", "manhattan"]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid = GridSearchCV(
    pipe,
    param_grid,
    cv=cv,
    scoring="accuracy",
    n_jobs=-1,
    verbose=1
)

# ------------------------
# 2) Entraînement
# ------------------------

grid.fit(X_train_scaled2, y_train2)

print("\nMeilleurs hyperparamètres :")
print(grid.best_params_)
print("\nBest CV Accuracy:", grid.best_score_)

# ------------------------
# 3) Évaluation finale
# ------------------------

best_model = grid.best_estimator_
y_pred = best_model.predict(X_test_scaled2)

acc = accuracy_score(y_test2, y_pred)
print(f"\nTest Accuracy: {acc*100:.2f}%")

print("\nClassification Report:")
print(classification_report(y_test2, y_pred, target_names=le.classes_))

# ------------------------
# 4) Matrice de confusion
# ------------------------

cm = confusion_matrix(y_test2, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d",
            xticklabels=le.classes_,
            yticklabels=le.classes_)
plt.title(f"Confusion Matrix (Accuracy={acc*100:.2f}%)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```
